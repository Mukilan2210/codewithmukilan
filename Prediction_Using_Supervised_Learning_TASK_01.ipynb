# **Predicting the percentage of a student based on the number of study hours.**

```
#Importing libraries for data manipulation and visualization
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#Importing sklearn library tools
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```
#importing dataset (Note : The dataset is directly used from the website itself)
dataset = pd.read_csv("http://bit.ly/w-data")
dataset.head()
```

```
#To identify the content in dataset
dataset.shape
```

```
dataset.describe() #To identify the statistical values from the dataset
```

```
#Plotting the dataset 
dataset.plot(x='Hours',y='Scores',style='o')
plt.title('Hours vs Percentage')
plt.xlabel('Hours studied')
plt.ylabel('Percentage obtained')
#To Illustrate the plotted dataset via graph
plt.show()
```

Since, We have obtained A rising change in the **percentile of marks** obtained by the students, we can come to the conclusion that the students have started preparing well in the **duecourse of Time (Hours)**.

This Can also be denoted as **Positive Linear Relation**.

```
#Selecting Dependent and Independent variables from the dataset 
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values

#Spliting the Dataset into training_set and test_set
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.4)
```

```
#Preparing a Recipe for linear Regression
reg = LinearRegression()
reg.fit(X_train, Y_train)
```

```
#Visualising the training_set dataset by applying Linear Regression
plt.scatter(X_train, Y_train, color = 'red')
plt.plot(X_train, reg.predict(X_train), color = 'blue')
plt.title('Hours vs Percentage')
plt.xlabel('Hours studied')
plt.ylabel('Percentage obtained')
plt.show()
```

```
#Predicting the Test_set dataset result
Y_pred = reg.predict(X_test)

#Visualising the Test_set Dataset using Linear Regression
plt.scatter(X_test, Y_test, color = 'red')
plt.plot(X_train, reg.predict(X_train), color = 'blue')
plt.title('Hours vs Percentage')
plt.xlabel('Hours studied')
plt.ylabel('Percentage obtained')
plt.show()
```

```
#Comparing Original and predicted values of test_set
df = pd.DataFrame({'Actual': Y_test,'Predicted': Y_pred})
df
```

```
#Model Evaluation
print('Mean Absolute Error :',metrics.mean_absolute_error(Y_test, Y_pred))
print('Mean Squared Error :',metrics.mean_squared_error(Y_test, Y_pred))
```

Now We have identified the M.A.E and M.S.E values of the dataset by using Linear Regression.

## **Calculating the Predicted Score if a student studies for 9.25 hrs/day?**

```
reg.predict([[9.25]])
```
